# Scraping Protein from RCSB PDB

```python
from petastic4.scrapping import rcsb_search, get_fasta
df = rcsb_search('petase')
df[['fasta_id', 'chain', 'sequence', 'fasta_desc']] = df['identifier'].apply(get_fasta)
df['seq_len'] = df['sequence'].apply(len)
df
```

```python
df.to_csv('../data/2025/petase.csv')
```
